```python
import matplotlib.pyplot as plt
import numpy as np

# Function to plot using a specific color cycle
def plot_with_color_cycle(color_cycle, title):
    plt.figure(figsize=(10, 6))
    plt.rcParams['axes.prop_cycle'] = plt.cycler(color=color_cycle)
    x = np.linspace(0, 10, 100)
    for i in range(len(color_cycle)):
        plt.plot(x, np.sin(x + i), label=f'Line {i+1}')
    plt.legend()
    plt.title(title)
    plt.show()

# Plot using 'tab10' color cycle
plot_with_color_cycle(plt.cm.tab10.colors, 'tab10 Color Cycle')

# Plot using 'tab20' color cycle
plot_with_color_cycle(plt.cm.tab20.colors, 'tab20 Color Cycle')

# Plot using 'tab20b' color cycle
plot_with_color_cycle(plt.cm.tab20b.colors, 'tab20b Color Cycle')

# Plot using 'tab20c' color cycle
plot_with_color_cycle(plt.cm.tab20c.colors, 'tab20c Color Cycle')

# Plot using 'Dark2' color cycle
plot_with_color_cycle(plt.cm.Dark2.colors, 'Dark2 Color Cycle')

# Plot using 'Paired' color cycle
plot_with_color_cycle(plt.cm.Paired.colors, 'Paired Color Cycle')

# Plot using 'Set1' color cycle
plot_with_color_cycle(plt.cm.Set1.colors, 'Set1 Color Cycle')

# Plot using 'Set2' color cycle
plot_with_color_cycle(plt.cm.Set2.colors, 'Set2 Color Cycle')

# Plot using 'Set3' color cycle
plot_with_color_cycle(plt.cm.Set3.colors, 'Set3 Color Cycle')

# Plot using 'Accent' color cycle
plot_with_color_cycle(plt.cm.Accent.colors, 'Accent Color Cycle')

# Plot using 'Pastel1' color cycle
plot_with_color_cycle(plt.cm.Pastel1.colors, 'Pastel1 Color Cycle')

# Plot using 'Pastel2' color cycle
plot_with_color_cycle(plt.cm.Pastel2.colors, 'Pastel2 Color Cycle')
```
